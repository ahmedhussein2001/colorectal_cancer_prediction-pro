Machine Learning 

```python
df = data[["age","bmi","stage_at_diagnosis", "tumor_aggressiveness", 
           "physical_activity_level", "smoking_status", "alcohol_consumption","time_to_diagnosis"]]
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

 1-    regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import root_mean_squared_error
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cat_val:
    df[col] = le.fit_transform(df[col]) 
df = df.copy()
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()  
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)  
```

```python
X = df.drop('time_to_diagnosis',axis=1)          # features
y = df['time_to_diagnosis']                      # target 
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=.25 , random_state=42 , shuffle=True)
```

Linear Regression


```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
y_pred = LR.predict(X_test)
```

```python
print(LR.score(X_train,y_train))
print(LR.score(X_test, y_test))
```

```python
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R Score:", r2_score(y_test, y_pred))
```

 Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y1_pred = rf.predict(X_test)
```

```python
print("MAE:", mean_absolute_error(y_test, y1_pred))
print("R Score:", r2_score(y_test, y1_pred))
```

Ridge Regression

```python
# Ridge Regression
from sklearn.linear_model import Ridge
rd = Ridge(alpha=0.5)

rd.fit(X_train,y_train)
y2_pred = rd.predict(X_test)

```

```python
print("MAE:", mean_absolute_error(y_test, y2_pred))
print("R Score:", r2_score(y_test, y2_pred))
```

  SGD Regressor

```python
from sklearn.linear_model import SGDRegressor
sdgr = SGDRegressor(alpha=50,random_state=1000,penalty='l2',shuffle=True)
sdgr.fit(X_train,y_train) 
```

```python
y3_pred = sdgr.predict(X_test)
```

```python
print("MAE:", mean_absolute_error(y_test, y3_pred))
print("R Score:", r2_score(y_test, y3_pred))
```
